```python
f = open('/home/lzy/lab/Project/TaihuData/svr_model/result/predictResult_d(t+3).json', 'r')
content = f.read()
predict = json.loads(content)
print(type(predict))
print(len(predict))
f.close()
```

```python
df = pd.read_csv('/home/lzy/lab/Project/TaihuData/Taihu3.csv')
dates = df.iloc[-566:,1]
groundtruth = df.iloc[-566:,2]
groundtruth.tolist()
dates
```

```python
ticker_spacing = groundtruth
ticker_spacing = 50
```

```python

#创建画布
# fig = plt.figure(figsize=(12, 4), dpi=200)
fig, ax = plt.subplots(1,1, figsize=(15, 10))
# plt.figure(figsize=(8, 10))
ax.plot(dates,groundtruth,predict)
ax.xaxis.set_major_locator(ticker.MultipleLocator(ticker_spacing))
plt.xticks(rotation = 45)

# 绘制第一个时间序列图
plt.plot(dates, predict, label='predict', color='blue')
# 绘制第二个时间序列图
plt.plot(dates, groundtruth, label='groundtruth', color='red')
plt.title('Multiple Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
